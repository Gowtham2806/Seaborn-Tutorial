```python
import seaborn as sns
import pandas as pd
```

```python
sales = pd.read_excel('sales.xlsx')
```

```python
sns.relplot(x='Order Date', y='Sales', data=sales, kind='line')
```

```python
sales['Year'].nunique()
```

```python
sns.relplot(x='Year', y='Sales', data=sales, kind='line')
```

```python
sns.relplot(x='Year', y='Sales', data=sales, kind='line', ci=None)
```

```python
sns.relplot(x='Year', y='Sales', data=sales, kind='line', ci='sd')
```

```python
sns.relplot(x='Year', y='Sales', data=sales, kind='line', estimator=None)
```

```python
sns.relplot(x='Year', y='Sales', data=sales, kind='line', hue='Ship Mode', estimator=None)
```

```python
sns.relplot(x='Year', y='Sales', data=sales, kind='line', hue='Ship Mode')
```

```python
sns.relplot(x='Year', y='Sales', data=sales, kind='line', hue='Ship Mode', ci=None)
```

```python
sns.relplot(x='Year', y='Sales', data=sales, kind='line', hue='Ship Mode', style='Product Category', ci=None)
```
