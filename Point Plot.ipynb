```python
import seaborn as sns
import pandas as pd
```

```python
sales = pd.read_excel('sales.xlsx')
```

```python
sns.catplot(x='Product Category', y='Sales', data=sales, kind='point')
```

```python
sns.catplot(x='Product Category', y='Sales', data=sales, kind='bar')
```

```python
sns.catplot(x='Product Category', y='Sales', data=sales, hue='Ship Mode', kind='point')
```

```python
sns.catplot(x='Product Category', y='Sales', data=sales, col='Ship Mode', row='Region', kind='point')
```
